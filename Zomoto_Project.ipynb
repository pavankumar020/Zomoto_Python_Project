# Using GroupBy to group (Aggregate rating, Rating color, Rating text)

```python
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

# Changing column name from '0' to 'Rating count'

```python
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

# Observations

```python
## When rating is between 4.5 to 4.9 it indicates ---> Excellent
## When rating is between 4.0 to 4.4 it indicates ---> Very Good
## When rating is between 3.5 to 3.9 it indicates ---> Good
## When rating is between 2.5 to 3.4 it indicates ---> Average
## When rating is between 1.8 to 2.4 it indicates ---> Poor
## When rating is 0.0 it indicates ---> Not rated
```

# Plot a graph of ratings with the help of barplot

```python
ratings
```

```python
import matplotlib

matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(data = ratings,x = 'Aggregate rating',y = 'Rating count' )
```

```python
sns.barplot(data = ratings,x = 'Aggregate rating',y = 'Rating count',hue = 'Rating color')
```

# Count plot
